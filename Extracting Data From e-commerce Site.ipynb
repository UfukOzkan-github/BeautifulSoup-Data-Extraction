```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python

#We create an empty list
#We will add the features to this list later
liste1 = []


#We take the first 15 pages, you can change them according to your wishes.
a = 1
while a<=15:
    
    
    #Here I had to use a user agent because the HTTP status code was 403.
    #If your status code is 200, you do not have to use a user agent.
    header={
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
    }
    url="https://www.hepsiburada.com/spor-ayakkabilar-c-384551?filtreler=cinsiyet%3AErkek&sayfa="+str(a)+""

    r=requests.get(url,headers=header)
    soup=BeautifulSoup(r.content,"lxml")
    
    #We start breaking down from source covering all products
    st1=soup.find("div",attrs={"class":"productListContent-pXUkO4iHa51o_17CBibU"})
    st2=st1.find("ul",attrs={"class":"productListContent-frGrtf5XrVXRwJ05HUfU productListContent-rEYj2_8SETJUeqNhyzSm"})
    st3=st2.find_all("li",attrs={"class":"productListContent-zAP0Y5msy8OHn5z7T_K_"})
    
    
    #In the for loop, we go to your products page one by one and get the properties.
    #Some products have the entire product link ("href")
    #while others do not have the entire link. We solve this situation with the if structure
    for detail in st3:
        top_of_link="https://www.hepsiburada.com"

        end_of_link=detail.a.get("href")
        if "https://adservice." not in end_of_link:
            link1=end_of_link = top_of_link + end_of_link

        else :
            link1=end_of_link


        print(link1) 
        
        
        #Here we use user agent again
        #Since the header variable is above, we do not write it again.
        url1=link1
        r1=requests.get(url1,headers=header)
        soup1=BeautifulSoup(r1.content,"lxml")
        
        #We pull the features we want to get one by one and assign them to a variable.
        new_price=soup1.find("span",attrs={"id":"offering-price"}).text.strip().replace("\n","")
        print(new_price)
        original_price=soup1.find("del",attrs={"id":"originalPrice"}).text.strip()
        print(original_price)
        discount=soup1.find("span",attrs={"id":"product-discount-rate"}).text.strip()
        print(discount)
        try:
            product_rating= soup1.find("span",attrs={"class":"rating-star"}).text.strip()
        except:
            product_rating = "puan yok"
        print(product_rating)
        try:
            number_of_reviews=soup1.find("div",attrs={"id":"comments-container"}).text.strip()

        except:
            number_of_reviews = "degerlendirme yok"
        print(number_of_reviews)
        
        name_of_the_product=soup1.find("h1",attrs={"id":"product-name"}).text.strip()
        print(name_of_the_product)
        brand_name=soup1.find("span",attrs={"class":"brand-name"}).text.strip()
        print(brand_name)
        product_photo=soup1.img.get("src")
        print(product_photo)
        numbers=soup1.find_all("div",attrs={"data-bind":"click: setActiveVariantContainer"})    
        shoe_num_liste=[]
        for num in numbers :
            shoe_num=num.find("label",attrs={"class":"label white"}).text.strip()
            print(shoe_num)
            shoe_num_liste.append(shoe_num)
        #We put the data we received into the empty list we created above.    
        liste1.append([name_of_the_product,brand_name,link1,product_photo,new_price,discount,original_price,product_rating,number_of_reviews,shoe_num_liste])
    #We increase the value of a to change the number of pages
    a = a+1
```

```python
#We increase the value of a to change the number of pages
df = pd.DataFrame(liste1)
df.columns=["name of the product","brand name","link"," product photo","new price","discount","original price","product rating","number of reviews","shoe num liste"]
```
